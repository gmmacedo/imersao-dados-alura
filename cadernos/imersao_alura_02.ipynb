```python
plt.figure(figsize=(9,7))
topcompdb = dados.query('composto in @topcomp')
ax = sns.countplot(x = 'composto', data = topcompdb, order= dados['composto'].value_counts().index[:5])
ax.set_title('Top 5 compostos')
ax.set_xlabel('Compostos')
ax.set_ylabel('Contagem')
plt.show()
```

##Desafio 03:

```python
hist1 = sns.histplot(data = dados, x= 'g0', bins=50)
hist1.set_title('Expressões do gene g0', fontsize=20)
hist1.set_xlabel('Valores de g0', fontsize=14)
hist1.set_ylabel('Contagem', fontsize=14)
plt.show()

hist2 = sns.histplot(data = dados, x= 'c0', bins=50)
hist2.set_title('Expressões do gene c0', fontsize=20)
hist2.set_xlabel('Valores de c0',fontsize=14)
hist2.set_ylabel('Contagem', fontsize=14)
plt.show()
```

###Desafio 04:

```python
d04 = dados.loc[:, 'g0' : 'g5']
d04.describe()
```

#### Neste 'describe()' podemos ver que foram criados oito linhas, cada uma tem um valor diferente. De forma decrescente vemos:

#### 'count': Significa a quantidade de valores que a coluna possui.
#### 'mean': Traduzindo, média, é a média aritimética dos valores da coluna.
#### 'std': Desvio padrão, é a taxa de variação ou dispersão dos valores da coluna.
#### 'min': É o menor valor contido na tabela.
#### '25%, 50%, 75%': O '50%' é a mediana, pu seja, colocando em ordem do menor para o maior é o valor que se encontra justamente no meio. '25%' é a porcentagem dos valores mais baixos e o '75%' é a porcentagem dos valores mais altos.
#### 'max': É o valor mais alto contido na coluna.

##Desafio 05:
####O tamanho do gráfico é essencial para a sua precisão, um gráfico muito pequeno pode causar em uma interpretação errada, pois algumas informações podem ser ocultadas ou de difícil localização. Isso fica mais notável em um gráfico com muitos pontos e principalmente quando há pequenas variações entre os pontos.

###Testando gráficos diferentes...

```python
plt.figure(figsize=(10,5))
h1 = sns.histplot(data = dados, x= 'g0', bins=50, fill=False, label='G0')
sns.histplot(data = dados, x= 'g123', bins=50, element='poly', label='G123')

h1.set_title('Comparação dos valores de g0 e g123', fontsize=20)
h1.set_xlabel('Valores de g0 e g123', fontsize=15)
h1.set_ylabel('Contagem', fontsize=15)
plt.show()
```

###Essa forma de fazer uma sobreposição dos gráficos talvez não seja a mais ideal, não me parece ter um suporte (talvez até seja um bug) pois tentar usar alguns parâmetros resulta em erro, atrapalhando até mesmo a interpretação dos dados, já que uma legenda é de essencial importância quando temos mais de uma variável em um gráfico.
