```python
dados.columns = dados.columns.str.replace('-','')
dados
```

```python
mapa={'droga': 'composto'}
dados.rename(columns= mapa, inplace=True)
```

```python
topcomp = dados['composto'].value_counts().index[:5]
topcomp
```

```python
dados.query('composto in @topcomp')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(9,7))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @topcomp'))
ax.set_title('Top 5 compostos')
plt.show()
```
